```python
import geopandas as gpd
import contextily as cx
import matplotlib.pyplot as plt
```

```python
# Corrigir coordenadas e salvar arquivo
pasta = "D:\\Downloads\\casan"
casan = gpd.read_file(pasta+"\\rede_casan.shp").set_crs('31982')
casan['geometry'] = casan.translate(xoff=-55, yoff=-45)
casan.to_file(pasta+"\\rede_casan_corrigido\\rede_casan_corrigido.shp") 
```

```python
# Visualizar 
fig, ax = plt.subplots(figsize=(10,10))
casan.plot(ax=ax, alpha=0.5, edgecolor="k")
cx.add_basemap(ax, crs=casan.crs)
ax.set_axis_off()
fig.tight_layout()
```
